1. #### Visualitza el nom dels mòduls de dificultat mitja.
S'ha de tenir esment de com construir les projeccions.
https://www.mongodb.com/docs/manual/reference/method/db.collection.find/#std-label-method-find-projection
```
db.moduls.find({"dificultat":"mitja"},{"nom":1})
```
Resultat:
>```json
{ _id: ObjectId("63738f58b8848879b1319385"), nom: 'SBD' }
{ _id: ObjectId("637399ddb8848879b1319387"), nom: 'BDA' }
```
1. #### Visualitza el nom dels mòduls amb més de trenta matriculats.
Els operadors de consultes i projeccions.
https://www.mongodb.com/docs/manual/reference/operator/query/
```
db.moduls.find({"matriculats":{$gte:30}},{"nom":1})
```
Resultat:
>```json
{ _id: ObjectId("63738f58b8848879b1319384"), nom: 'SAA' }
{ _id: ObjectId("63738f58b8848879b1319385"), nom: 'SBD' }
{ _id: ObjectId("63738f58b8848879b1319386"), nom: 'MIA' }
```
